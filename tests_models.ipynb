```python
import json
import pandas as pd
import numpy as np 
import pickle
```

```python
df = pd.read_csv("excel/data.csv")
json = json.load(open("json/labelParamsHP.json", "r"))
XGboosHP_v2 = pickle.load(open(f".\model\XGboosHP_v2.pkl","rb"))
```

```python
datos = []

for index, data in df.iterrows():
    hp = data["hp"]
    consulta = {
            "area_b": data["area_b"],
            "n_rtu": data["n_rtu"],
            "area_r": data["area_r"],
            "perimeter": data["perimeter"]
        }
    predict = XGboosHP_v2.predict([list(consulta.values())])
    consulta["hp"] = hp
    consulta["hp_predict_ls"] = predict[0]
    consulta["is_equal"] = hp == predict[0]
    
    datos.append(consulta)
    
```

```python
df = pd.DataFrame(datos)

# medimos el porcentaje de asierto del producto
cont = 0
for i in df["is_equal"]:
    if i: cont += 1
print(cont * 100 / len(df["is_equal"]))
```
